```python
import pandas as pd
import numpy as np
import math
import random
```

```python
df=pd.read_csv("16P.csv",encoding="ISO-8859-1") #use encoding for avoid from errors
df
```

```python
types = np.unique(np.array(df['Personality']))
types
```

```python
df.drop("Response Id",axis=1,inplace=True) #drop the unnecessary column
```

```python
df["Personality"]=df["Personality"].str.replace("ESTJ","0")
df["Personality"]=df["Personality"].str.replace("ENTJ","1")
df["Personality"]=df["Personality"].str.replace("ESFJ","2")
df["Personality"]=df["Personality"].str.replace("ENFJ","3")
df["Personality"]=df["Personality"].str.replace("ISTJ","4")
df["Personality"]=df["Personality"].str.replace("ISFJ","5")
df["Personality"]=df["Personality"].str.replace("INTJ","6")
df["Personality"]=df["Personality"].str.replace("INFJ","7")
df["Personality"]=df["Personality"].str.replace("ESTP","8")
df["Personality"]=df["Personality"].str.replace("ESFP","9")
df["Personality"]=df["Personality"].str.replace("ENTP","10")
df["Personality"]=df["Personality"].str.replace("ENFP","11")
df["Personality"]=df["Personality"].str.replace("ISTP","12")
df["Personality"]=df["Personality"].str.replace("ISFP","13")
df["Personality"]=df["Personality"].str.replace("INTP","14")
df["Personality"]=df["Personality"].str.replace("INFP","15")
```

```python
df
```

```python
df_normalize=df.copy()  #create a normalize function for get values between 0 and 1
def normalization(row):
    minimum=min(row)
    maximum=max(row)
    value=(row-minimum)/(maximum-minimum)
    return value
for column in df_normalize.columns[:-1]:
    arr=np.array(df_normalize[column])
    df_normalize[column]=normalization(arr)
df_normalize
```

```python
df_normalize = df_normalize.sample(frac=1) #use sample function for shuffle the data
X = np.array(df_normalize.drop(["Personality"],axis=1)) #decide dependet variable
y = np.array(df_normalize["Personality"].astype("int")) #decide independent variable
testx = list(np.array_split(X,5)) #split the data for take each fold
testy = list(np.array_split(y,5))
X_test = testx[0] #test data is one of this split
y_test = testy[0]
testx.pop(0)  #use pop for elimine test data from all data
testy.pop(0)
X_train = np.concatenate(testx) #train data is rest of this slpit
y_train = np.concatenate(testy)
```

```python
def distances(X_test, X_train): #create distance function for calculate euqlidead distance of each row
    distance = -2 * np.dot(X_test, X_train.T) + np.sum(X_train**2,    axis=1) + np.sum(X_test**2, axis=1)[:, np.newaxis]
    distance=np.sqrt(distance)
    return distance

distance=distances(X_test,X_train)
print(distance)
```

```python
def K_Neighbors(train, test, num): #knn algorith for find the nearest neighbor i sort my list in this function for find the 
    neighbors=[]                   #prediction at the same time, it will return prediction
    distance=distances(test,train)
    sorted_dist=distance.argsort()
    for dist in sorted_dist:
        data  = y_train[dist]
        neighbors.append(data[:num])
    return neighbors
K_Neighbors(X_train, X_test, 1)
```

```python
prediction = K_Neighbors(X_train, X_test, 1)
def confusion_matrix(true, pred): #confusion matrix for find tp,tn,fp,fn values

  K = len(np.unique(true)) # Number of classes 
  result = np.zeros((K, K)) #zeros matrix

  for i in range(len(true)): #implement our prediction in confusion matrix
    result[true[i]][pred[i]] += 1

  return result
print(confusion_matrix(y_test,prediction))
```

```python
matrix=confusion_matrix(y_test,prediction)
def evaluate(matrix,k): #create a evulate function for calculate aur percentage acuarcy,recall and precision values
    accuracy = 0        # by using confusion matrix for this calculations
    precision = 0
    recall = 0
    for i in range(np.shape(matrix)[0]):
        tp = matrix[i, i]
        fp = np.sum(matrix[i]) - tp
        fn = np.sum(matrix[:, i]) - tp
        tn = np.sum(matrix) - tp - fp - fn
        acc = (tp + tn) / (tp + fp + fn + tn)
        accuracy += acc
        prec = tp / (tp + fp)
        precision += prec
        rec = tp / (tp + fn)
        recall += rec
    print("\nFor k: {} ".format(k))
    print("Accuracy: %", round((accuracy / 16) * 100, 2),end="")
    print(" Precision: %", round((precision / 16) * 100, 2),end="")
    print(" Recall: %", round((recall / 16) * 100, 2),end="")
evaluate(matrix,1)
```

```python
for i in range(5): # calculate this value for each fold
    X = np.array(df_normalize.drop(["Personality"],axis=1))
    y = np.array(df_normalize["Personality"].astype("int"))
    testx = list(np.array_split(X,5))
    testy = list(np.array_split(y,5))
    X_test = testx[i]
    y_test = testy[i]
    testx.pop(i)
    testy.pop(i)
    X_train = np.concatenate(testx)
    y_train = np.concatenate(testy)
    k_list=[1,3,5,7,9] #calculate this value for each k value
    print("\n\nFold {}".format(i+1))
    for k in k_list:
        prediction = K_Neighbors(X_train, X_test, k)
        matrix=confusion_matrix(y_test,prediction)
        evaluate(matrix,k)
    
```
